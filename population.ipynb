```python
import numpy as np
import pandas as pd
import folium as fo 
%matplotlib inline
```

```python
map = fo.Map()
map
```

```python
popu=pd.read_csv('in.csv')
```

```python
popu.head()
```

```python
lat_po = list(popu['lat'])
lng_po = list(popu['lng'])
city_po = list(popu['city'])
pop_po = list(popu['population'])
```

```python
po = fo.FeatureGroup(name='My Map')
```

```python
def mar(popu):
    if (popu>10000000):
        return 'red'
    elif (popu>5000000 and popu<=10000000):
        return 'blue'
    else:
        return 'green'
```

```python
for lat,lng,city,pop in zip(lat_po,lng_po,city_po,pop_po):
    po.add_child(fo.Marker(location=[lat,lng],popup=[city,pop],icon=fo.Icon(color=mar(pop))))
```

```python
map.add_child(po)
```
